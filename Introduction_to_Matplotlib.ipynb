Pylot API vs object_Oriented (OO) API


pylot API

```python
x = np.linspace(0,100,1000)
plt.plot(x, np.sin(x), label="Sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.xlabel("Bien X")
plt.ylabel("F(x)")
plt.title("sin(x) and cos(x)")
plt.legend()
```

OO_API

```python
x = [1,2,3,4]
y = [11,22,33,44]
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y);
ax.set(title="A simple plot",
        xlabel="x-axis",
        ylabel="y-axis");
```

Project : Data Visualzation - California cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./data/california_cities.csv')
cities.head()
```

```python
# tach vi do va kinh do
lat, lon = cities['latd'], cities['longd']
lat[:10]

#plot using Pylot API

plt.scatter(lon,lat)
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude')
```

```python
population , area = cities['population_total'], cities['area_total_km2']
# tach vi do va kinh do
lat, lon = cities['latd'], cities['longd']
lat[:10]

#plot using Pylot API

plt.scatter(lon,lat,c=np.log10(population),cmap='viridis')
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)
```
